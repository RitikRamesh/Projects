```python
# label encoding

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['airline']=le.fit_transform(df['airline'])
df['source_city']=le.fit_transform(df['source_city'])
df['departure_time']=le.fit_transform(df['departure_time'])
df['stops']=le.fit_transform(df['stops'])
df['arrival_time']=le.fit_transform(df['arrival_time'])
df['destination_city']=le.fit_transform(df['destination_city'])
df['class']=le.fit_transform(df['class'])
```

```python
# Feature selection


plt.figure(figsize=(10,5))
sns.heatmap(df.corr(),annot=True,cmap='coolwarm')
plt.show()
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
col_list = []
for col in df.columns:
    if ((df[col].dtype != 'object') & (col != 'price')):
        col_list.append(col)
        
X = df[col_list]
vif_data = pd.DataFrame()
vif_data['feature'] = X.columns
vif_data['VIF'] = [variance_inflation_factor(X.values, i)
                   for i in range(len(X.columns))]
vif_data
```

```python
df = df.drop(columns=['stops'])

col_list = []
for col in df.columns:
    if ((df[col].dtype != 'object') & (col != 'price')):
        col_list.append(col)
        
X = df[col_list]
vif_data = pd.DataFrame()
vif_data['feature'] = X.columns
vif_data['VIF'] = [variance_inflation_factor(X.values, i)
                   for i in range(len(X.columns))]
vif_data
```

```python
# Linear Regression model

# 1. Split the data 
X = df.drop(columns=['price','flight'])
Y = df['price']

from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
# 2. Preprocessing
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
# 3. Build Model
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train,Y_train)
```

```python
# 4. Predict
y_pred = regressor.predict(X_test)
```

```python
# 5. Evaluate
from sklearn.metrics import r2_score
r2_score(Y_test,y_pred)
```

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(Y_test,y_pred)
```

```python
from sklearn.metrics import mean_absolute_percentage_error
mean_absolute_percentage_error(Y_test,y_pred)
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(Y_test,y_pred)
```

```python
root_mean_square_error = np.sqrt(mean_squared_error(Y_test,y_pred))
root_mean_square_error
```

```python
# 6. Visualize
plt.scatter(Y_test,y_pred)
plt.xlabel('Y Test')
plt.ylabel('Predicted Y')
plt.title('Linear Regression')
plt.show()
```

```python
sns.displot(Y_test,kde=True,rug=True,label='Actual')
sns.displot(y_pred,kde=True,rug=True,label='Predicted')
plt.legend()
plt.show()
```
